# Question #2

```python
class Student:
    def __init__(self, name, fsu_id, courses, target_graduation_year):
        self.name = name
        self.id = fsu_id
        self.course = courses
        self.graduation = target_graduation_year
           
    def __str__(self):
        return f"Student's name: {self.name}\nFSU ID: {self.id}\nCourses: {self.course} \nTarget Graduation: {self.graduation}"
    
    def add_course(self, courseName):
        self.course = self.course + ", " + courseName
        
        
    def years_left_till_graduation(self):
        self.years = self.graduation - 2019
        return self.years
        
student_1 = Student('Terrell', 12345678, 'ISC-3313', 2022)
student_2 = Student('Marshall', 87654321, 'ENC-1102', 2024)

print(student_1)
print("\n")
student_1.add_course('MAS-3105')
print(student_1)

print(f"Years left: {student_1.years_left_till_graduation()} years")
print("\n")

student_2.add_course('COP-3330')
print(student_2)
print(f"Years left: {student_2.years_left_till_graduation()} years")
```

```python
ax = plt.axes()

t = np.linspace(0, 2 * np.pi, 1000)

a = 3
b = 2

delta = np.pi / 2

A = B = 1

x = (A * np.sin(a*t + delta))

y = (B*np.sin(b*t))

ax.plot(x, y, color = 'olive')
```

# Part B

```python
fig, ax = plt.subplots(4, 4,) #creating a 4x4 figure

a = [1, 2, 3, 4] #list a
b = [1, 2, 3, 4] #list b

for i in range(0,4): #controls the rows
    for j in range(0,4): #controls the columns
        x = A*np.sin(a[i]*t + delta)
        y = B*np.sin(b[j]*t)
        ax[i,j].plot(x,y, color = 'violet')
        ax[i,j].axis("equal")
```

```python
import numpy as np
class Vector:
    def __init__(self, a=0, b=0, c=0):
        self.a = a
        self.b = b
        self.c = c
    
    def __str__(self):
        return f"{self.a}*i + {self.b}*j + {self.c}*k"
        
    def __add__(self, other):
        return Vector(self.a + other.a, self.b + other.b, self.c + other.c)
        
    def __sub__(self, other):
        return Vector(self.a - other.a, self.b - other.b, self.c - other.c)

    def dot(self, other):
        if isinstance(other, Vector): # dot product method
            return self.a * other.a + self.b * other.b + self.c * other.c
        
    def cross(self, other):
        if isinstance(other, Vector): # cross multiplying method
            return Vector(self.b * other.c - other.b *self.c, other.a * self.c - self.a - other.c, self.a * other.b - other.a * self.b)
    
    def distance(self, other):
        da = self.a - other.a
        db = self.b - other.b
        dc = self.c - other.c
        return np.sqrt(da**2 + db**2 + dc**2)
    
    def __abs__(self):
        return self.distance(Vector())
    
#Vectors 1 and 2    
x = Vector(4,5,6)
y = Vector(7,8,9)

# Adding two vectors
v1 = x + y
print(f"Adding the two vectors we get: {v1}")

#Subtracting two vectors
v2 = x - y
print(f"\nSubtracting the two vectors we get: {v2}")

#Doing the dot product of two vectors
v3 = x.dot(y)
print(f"\nThe dot product results: {v3}")

#Cross multiplying two vectors
v4 = x.cross(y)
print(f"\nCross multiplying results: {v4}")

#Finding the magnitude of two vectors
v5 = abs(x)
print(f"\nThe magnitude is: {v5}")

#Conversion to string 
v6 = str(x)
print(f"\nString conversion: {v6}")

```
